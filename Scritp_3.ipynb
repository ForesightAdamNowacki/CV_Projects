## 3. Reading, Transforming and Exporting Videos using OpenCV and Python

```python
import cv2
import os
import matplotlib.pyplot as plt

os.chdir(r"D:\GitHub\Github_CV_Projects")

video_id = 0
```

## Read Video

```python
# Capture video:
cap = cv2.VideoCapture(os.path.join('Recordings','Sign_Language', os.listdir(os.path.join('Recordings','Sign_Language'))[video_id]))
# Grab a frame
cap, frame = cap.read()
# Render frame 
plt.imshow(cv2.cvtColor(frame, cv2.COLOR_BGR2RGB))
plt.show()
```

## Capture properties:

```python
# Capture video:
cap = cv2.VideoCapture(os.path.join('Recordings','Sign_Language', os.listdir(os.path.join('Recordings','Sign_Language'))[video_id]))
# Height
print(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
# Width
print(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
# Get number of frames
print(cap.get(cv2.CAP_PROP_FRAME_COUNT))
# Frames per second
print(cap.get(cv2.CAP_PROP_FPS))
# Video time:
print(cap.get(cv2.CAP_PROP_FRAME_COUNT)/cap.get(cv2.CAP_PROP_FPS), "seconds")
```

## Working with Video Capture

```python
# Establish capture
cap = cv2.VideoCapture(os.path.join('Recordings','Sign_Language', os.listdir(os.path.join('Recordings','Sign_Language'))[video_id]))
# Loop through each frame
for frame_idx in range(int(cap.get(cv2.CAP_PROP_FRAME_COUNT))):
    # Read frame 
    ret, frame = cap.read()
    # Gray transform
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    # Show image
    cv2.imshow('Video Player', gray)
    # Breaking out of the loop
    if cv2.waitKey(10) & 0xFF == ord('q'):
        break

# Close down everything
cap.release()
cv2.destroyAllWindows()
```

## Save gray video

```python
# Establish capture
cap = cv2.VideoCapture(os.path.join('Recordings','Sign_Language', os.listdir(os.path.join('Recordings','Sign_Language'))[video_id]))
# Properties
height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
fps = cap.get(cv2.CAP_PROP_FPS)
# Video Writer 
video_writer = cv2.VideoWriter(os.path.join('Recordings','Sign_Language', "output_video_" + str(video_id) + ".avi"), cv2.VideoWriter_fourcc('P','I','M','1'), fps, (width, height), isColor = False) 
                                                                                        
# Loop through each frame
for frame_idx in range(int(cap.get(cv2.CAP_PROP_FRAME_COUNT))):
    # Read frame 
    ret, frame = cap.read()
    # Gray transform
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    # Show image
    cv2.imshow('Video Player', gray)
    # Write out frame 
    video_writer.write(gray)
    # Breaking out of the loop
    if cv2.waitKey(10) & 0xFF == ord('q'):
        break

# Close down everything
cap.release()
cv2.destroyAllWindows()
# Release video writer
video_writer.release()
```
